## Limpieza de datos

Me dí cuenta que la mayor parte de las columnas tienen este signo '/r' porque la información esta en csv y se separa a partir de cambios de columna. Esto dificulta el análisis de datos, por lo que es necesario limpiarlo

```python
# estoy quitando el '/r' solo a las columnas de texto una columna a la vez
# los reemplaza por vacios y de una vez quita los espacios al principio y al final
df_exploratorio = df_exploratorio.apply(
                lambda col: 
                col.str.replace('\r', '', regex=False).str.strip() 
                      if col.dtype == 'object' else col)
```

```python
# vemos si quedó bien 
df_exploratorio.head()
```

## Privatización

De acuerdo con los datos del  Informe sobre la situación de las prisiones privadas en México (Documenta y México Evalua, 2021) existen solamente 8 Centros Federales de Reinserción Social (Cefereso) que funcionan bajo la figura de Centro de Prestación de Servicios (CPPS).
Estos son: 
- Cefereso núm. 11 CPS Sonora
- Cefereso núm. 12 CPS Guanajuato
- Cefereso núm. 13 CPS Oaxaca
- Cefereso núm. 14 CPS Durango
- Cefereso núm. 15 CPS Chiapas
- Cefereso núm. 16 CPS Femenil Morelos
- Cefereso núm. 17 CPS Michoacán
- Centro Penitenciario Federal núm. 18 CPS Coahuila

En la encuesta hay muchos datos vacios, por lo que en pirmera instancia es importante separar los datos de fuero federal que contienen los Cefereso

```python
# Revisar si no hay datos vacios en esta columna
df_exploratorio['FUERO'].value_counts()
```

De acuerdo con el catálogo de esta columna, el número 2 es para centros estatales o municipales y el 1 para federales 

```python
# Nuevo df donde solo existan los datos de centros federales 
df_federal=df_exploratorio[df_exploratorio['FUERO']==1].copy()
df_federal.head()
```

```python
# conteo de con cuantos datos estamos trabajando
df_federal.shape
```

Ahora, sabemos los estados donde se encuentran los centros que funcionan bajo la figura del CPS

```python
# nuevo df con los estados que queremos 
estados_cps = ['Oaxaca','Sonora', 'Guanajuato', 'Durango', 'Chiapas', 'Morelos', 'Michoacán de Ocampo', 'Coahuila de Zaragoza']
df_estados=df_federal[df_federal['NOM_ENT'].isin(estados_cps)].copy()
df_estados.head()
```

```python
# Ahora hay que ver que centros federeales se encuentran en los estados
for ent, grupo in df_estados.groupby('NOM_ENT'):
    print(f"\n{ent}:")
    print(grupo['NOM_INT'].unique())
```

Hay datos vacios en Chiapas, Durango, Michoacán y Morelos. De acuerdo con los datos del [Observatorio de Prisiones](https://observatorio-de-prisiones.documenta.org.mx/category/centros-penitenciarios/) solo existe un centro federal en Chiapas y Michoacán, sin embargo en Morelos existe el Centro Federal de Rehabilitación Psicosocial (CEFERPSI) y en Durango el CEFERESO No 7. Nor-Noroeste Durango. 

```python
# Para asegurarnos que no esten en los datos originales checamos los valores 
df_exploratorio['NOM_INT'].unique()
```

No existe en los nombres de los centros, por lo que es posible asumir que pueden ser los dats que faltan. Es por eso que vamos a excluirlos del df

```python
#Establecemos una condición y la aplicamos al df para Durango
condicion_vacia = (df_estados['NOM_ENT'] == 'Durango') & (df_estados['NOM_INT']== '')
df_estados= df_estados[~condicion_vacia]
df_estados.head()
```

```python
#Establecemos una condición y la aplicamos al df para Morelos
condicion_vacia = (df_estados['NOM_ENT'] == 'Morelos') & (df_estados['NOM_INT']== '')
df_estados= df_estados[~condicion_vacia]
```

Como para Chiapas y Michoacan solo hay un solo centro federal, podemos asumir que estan hablando del CEFERESO No. 15 CPS Chiapas y el CEFERESO No. 17 CPS Michoacán  por lo que vamos a añadir esta información 

```python
df_estados['Chiapas'] = df_estados['Chiapas'].replace('', 'CEFERESO 15 CPS CHIAPAS')
df_estados['Chiapas']
```
