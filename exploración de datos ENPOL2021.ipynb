## Limpieza de datos

Me dÃ­ cuenta que la mayor parte de las columnas tienen este signo '/r' porque la informaciÃ³n esta en csv y se separa a partir de cambios de columna. Esto dificulta el anÃ¡lisis de datos, por lo que es necesario limpiarlo

```python
# estoy quitando el '/r' solo a las columnas de texto una columna a la vez
# los reemplaza por vacios y de una vez quita los espacios al principio y al final
df_exploratorio = df_exploratorio.apply(
                lambda col: 
                col.str.replace('\r', '', regex=False).str.strip() 
                      if col.dtype == 'object' else col)
```

## PrivatizaciÃ³n

De acuerdo con los datos del  Informe sobre la situaciÃ³n de las prisiones privadas en MÃ©xico (Documenta y MÃ©xico Evalua, 2021) existen solamente 8 Centros Federales de ReinserciÃ³n Social (Cefereso) que funcionan bajo la figura de Centro de PrestaciÃ³n de Servicios (CPPS).
Estos son: 
- Cefereso nÃºm. 11 CPS Sonora
- Cefereso nÃºm. 12 CPS Guanajuato
- Cefereso nÃºm. 13 CPS Oaxaca
- Cefereso nÃºm. 14 CPS Durango
- Cefereso nÃºm. 15 CPS Chiapas
- Cefereso nÃºm. 16 CPS Femenil Morelos
- Cefereso nÃºm. 17 CPS MichoacÃ¡n
- Centro Penitenciario Federal nÃºm. 18 CPS Coahuila

En la encuesta hay muchos datos vacios, por lo que en pirmera instancia es importante separar los datos de fuero federal que contienen los Cefereso

```python
# Revisar si no hay datos vacios en esta columna
df_exploratorio['FUERO'].value_counts()
```

De acuerdo con el catÃ¡logo de esta columna, el nÃºmero 2 es para centros estatales o municipales y el 1 para federales 

```python
# Nuevo df donde solo existan los datos de centros federales 
df_federal=df_exploratorio[df_exploratorio['FUERO']==1].copy()
df_federal.head()
```

```python
# conteo de con cuantos datos estamos trabajando
df_federal.shape
```

Ahora, sabemos los estados donde se encuentran los centros que funcionan bajo la figura del CPS

```python
# nuevo df con los estados que queremos 
estados_cps = ['Oaxaca','Sonora', 'Guanajuato', 'Durango', 'Chiapas', 'Morelos', 'MichoacÃ¡n de Ocampo', 'Coahuila de Zaragoza']
df_estados=df_federal[df_federal['NOM_ENT'].isin(estados_cps)].copy()
df_estados.head()
```

```python
# Ahora hay que ver que centros federeales se encuentran en los estados
for ent, grupo in df_estados.groupby('NOM_ENT'):
    print(f"\n{ent}:")
    print(grupo['NOM_INT'].unique())
```

Hay datos vacios en Chiapas, Durango, MichoacÃ¡n y Morelos. De acuerdo con los datos del [Observatorio de Prisiones](https://observatorio-de-prisiones.documenta.org.mx/category/centros-penitenciarios/) solo existe un centro federal en Chiapas y MichoacÃ¡n, sin embargo en Morelos existe el Centro Federal de RehabilitaciÃ³n Psicosocial (CEFERPSI) y en Durango el CEFERESO No 7. Nor-Noroeste Durango. 

```python
# Para asegurarnos que no esten en los datos originales checamos los valores 
df_exploratorio['NOM_INT'].unique()
```

No existe en los nombres de los centros, por lo que es posible asumir que pueden ser los dats que faltan. Es por eso que vamos a excluirlos del df

```python
# Checar si la condiciÃ³n se construyo bien 
condicion_vacia_D = (df_estados['NOM_ENT'] == 'Durango') & (df_estados['NOM_INT']== '')
condicion_vacia_D.value_counts()
```

```python
#Establecemos una condiciÃ³n y la aplicamos al df para Durango
df_estados= df_estados[~condicion_vacia_D]
df_estados.head()
```

```python
condicion_vacia_M = (df_estados['NOM_ENT'] == 'Morelos') & (df_estados['NOM_INT']== '')
condicion_vacia_M.value_counts()
```

```python
#Establecemos una condiciÃ³n y la aplicamos al df para Morelos
df_estados= df_estados[~condicion_vacia_M]
df_estados.head()
```

Como para Chiapas y Michoacan solo hay un solo centro federal, podemos asumir que estan hablando del CEFERESO No. 15 CPS Chiapas y el CEFERESO No. 17 CPS MichoacÃ¡n  por lo que vamos a aÃ±adir esta informaciÃ³n 

```python
condicion_vacia_C = (df_estados['NOM_ENT'] == 'Chiapas') & (df_estados['NOM_INT']== '')
condicion_vacia_C.value_counts()
```

```python
df_estados.loc[condicion_vacia_C, 'NOM_INT'] = 'CEFERESO 15 CPS CHIAPAS'
df_estados.head()
```

```python
condicion_vacia_M = (df_estados['NOM_ENT'] == 'MichoacÃ¡n de Ocampo') & (df_estados['NOM_INT']== '')
condicion_vacia_M.value_counts()
```

```python
df_estados.loc[condicion_vacia_M, 'NOM_INT'] = 'CEFERESO No. 17 CPS MichoacÃ¡n'
df_estados.head()
```

```python
url_estados = "https://geodata.ucdavis.edu/gadm/gadm4.1/json/gadm41_MEX_1.json"
estados = gpd.read_file(url_estados)
```

Primero hay que revisar como estan escritos los estados en el gdt

```python
print(estados["NAME_1"].sort_values().tolist())
```

Sabemos que Coahuila estÃ¡ como Coahuila de Zaragoza y MichoacÃ¡n como MichoacÃ¡n de Ocampo, por lo que hay que homologizarlos

```python
Datos_mapa ['NOM_ENT']= Datos_mapa ['NOM_ENT'].replace ('Coahuila de Zaragoza','Coahuila')
Datos_mapa
```

```python
Datos_mapa ['NOM_ENT']= Datos_mapa ['NOM_ENT'].replace ('MichoacÃ¡n de Ocampo','MichoacÃ¡n')
Datos_mapa
```

Establecemos la condiciÃ³n para iluminar los estados con ceresos privatizados

```python
estados['cfrs']= estados["NAME_1"].isin(Datos_mapa["NOM_ENT"])
estados['cfrs']
```

```python
#graficamos la informaciÃ³n

fig, ax = plt.subplots(figsize=(14, 10))
estados[estados['cfrs']].plot(
    ax=ax, color='#2bb6c0', edgecolor='#1a1919', linewidth=0.8
)
estados[~estados['cfrs']].plot(
    ax=ax, color='#e8e8e8', edgecolor='#1a1919', linewidth=0.8
)
parche_con = mpatches.Patch(color='#2bb6c0', label='Estado con CEFERESO bajo CPS')
# config. mapa 
ax.legend(handles=[parche_con], loc='lower left', fontsize=10)
ax.set_title(
    'Centros Federales de ReadaptaciÃ³n Social (CEFERESOS)\nque funcionan bajo la figura de Centro de PrestaciÃ³n de Servicios (CPPS)\nMÃ©xico',
    fontsize=14, fontweight='bold', pad=14
)
# agregar estados
for _, row in estados[estados['cfrs']].iterrows():
    centroid = row.geometry.centroid
    ax.annotate(
        row['NAME_1'],
        xy=(centroid.x, centroid.y),
        ha='center', va='center',
        fontsize=7.5, color='white', fontweight='bold'
    )
ax.set_axis_off()
plt.tight_layout()
# plt.savefig("ceferesos_cps.png", dpi=150, bbox_inches="tight")
plt.show()
```

```python
vmax = max(mapa['conteo'].max() for mapa, _, _ in centros_info)
vmin = 0
```

```python
fig, axes = plt.subplots(2, 4, figsize=(24, 12))
axes = axes.flatten()

for i, (mapa, estado, titulo) in enumerate(centros_info):
    ax = axes[i]

    mapa.plot(
        ax=ax,
        column='conteo',
        cmap='YlOrRd',
        edgecolor='#aaaaaa',
        linewidth=0.4,
        missing_kwds={'color': '#e8e8e8'},
        legend=False,
        vmin=vmin,
        vmax=vmax,
    )

    mapa[mapa['NAME_1'] == estado].plot(
        ax=ax, color='none', edgecolor='#2bb6c0', linewidth=2, hatch='///'
    )

    ax.set_title(titulo, fontsize=7, fontweight='bold', pad=6)
    ax.set_axis_off()
sm = plt.cm.ScalarMappable(cmap='YlOrRd', norm=plt.Normalize(vmin=vmin, vmax=vmax))
sm.set_array([])
fig.colorbar(sm, ax=axes, orientation='vertical', fraction=0.02, pad=0.02, label='NÃºmero de personas')

parche_cfrs = mpatches.Patch(facecolor='none', edgecolor='#2bb6c0', hatch='///', label='Estado del centro')
fig.legend(handles=[parche_cfrs], loc='lower center', fontsize=10, bbox_to_anchor=(0.5, 0.01))

fig.suptitle('Estado de origen de personas por CEFERESO CPS', fontsize=16, fontweight='bold', y=1.01)
plt.subplots_adjust(wspace=0.05, hspace=0.1)
plt.savefig('grid_ceferesos.png', dpi=150, bbox_inches='tight')
plt.show()
```
